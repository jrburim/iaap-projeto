# 4. Visualizando o resultado

```python
from IPython.display import HTML
from base64 import b64encode
import os

def mostrar_video(file_name):
      # Caminho do vídeo de entrada
      caminho_salvar = f'/content/iaap-projeto/data/results/{file_name}.mp4'

      # Caminho do vídeo comprimido
      caminho_comprimido = "/content/result_compressed.mp4"

      # Comprime o vídeo usando o ffmpeg
      os.system(f"ffmpeg -i {caminho_salvar} -vcodec libx264 {caminho_comprimido}")

      # Mostra o vídeo comprimido
      mp4 = open(caminho_comprimido,'rb').read()
      data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
      return HTML("""
      <video width=400 controls>
            <source src="%s" type="video/mp4">
      </video>
      """ % data_url)

mostrar_video('test_bytetrack.mp4')
```
